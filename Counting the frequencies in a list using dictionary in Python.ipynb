```python
input1 = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,
                  4, 4, 4, 2, 2, 2, 2]
input1
```

```python
freq = {}
for i in input1:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
for key,value in freq.items():
    print(key,':',value)
```

```python
#Using Count Functions
freq = {}
for items in input1:
    freq[items] = input1.count(items)
for key,value in freq.items():
    print(key,':',value)
```

```python
#Using Count.get functions
freq = {}
for items in input1:
    freq[items] = freq.get(items, 0) + 1
for key,value in freq.items():
    print(key,':',value)
```
