```python
input1 = {'a':100,'b':200,'c':300}
sums = 0
for i in input1:
    sums = sums + input1[i]
print("The sums of items in the dict:",sums)
```

```python
#Using Values() functions:
sums = 0
for i in input1.values():
    sums = sums + i 
print("The sums of items in the dict:",sums)
```
